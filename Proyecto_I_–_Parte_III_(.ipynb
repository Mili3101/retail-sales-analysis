# Carga de Datos

```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/Data Science - Skillnet/Ejercicios Core/BBDD/retail_sales_dataset.csv"
df = pd.read_csv(path)
```

# Limpieza de datos

```python
df.info()
```

```python
df.head(10)
```

```python
#se cambian espacios por guion bajo y se eliminan mayúsculas.
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
# Convertir la columna Date a datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

# Transformación de datos

```python
#Crea una columna que clasifique las ventas en categorías significativas (e.g., ‘Alta’, ‘Media’, ‘Baja’).

clase_ventas = [0, df['total_amount'].quantile(0.3), df['total_amount'].quantile(0.7), df['total_amount'].max()]
label_ventas = ['Baja', 'Media', 'Alta']
df['clase_ventas'] = pd.cut(df['total_amount'], bins=clase_ventas, labels=label_ventas)
```

```python
df[df['clase_ventas']== 'Baja']
```

```python
#Clasifica las ventas por categoria de producto
ventas_categorias = df.groupby('product_category')['total_amount']
```

```python
ventas_categorias.value_counts()
```

```python
#Clasifica las ventas por grupo de edades

ventas_edad = df.groupby('age')['total_amount']
```

```python
ventas_edad.value_counts()
```

```python
 #Ordena las ventas por genero y tipo de producto
ventas_generos= df.groupby(['product_category', 'gender'])['total_amount'].agg(
    ventas_totales='sum',
    ventas_prom='mean',
    cuenta_transact='count'
).reset_index()
```

```python
ventas_generos.value_counts()
```

```python
#Avanzadas
# Porcentaje del total por categoría
category_porcentaje = df.groupby('product_category')['total_amount'].sum() / df['total_amount'].sum() * 100
```

```python
category_porcentaje
```

```python
# Comparación entre géneros
gender_comparison = df.pivot_table(
    index='product_category',
    columns='gender',
    values='total_amount',
    aggfunc=['sum', 'mean']
)
```

```python
gender_comparison
```

```python
# Top 3 categorías de producto por ventas totales
top_categorias = df.groupby('product_category')['total_amount'].sum()
top_categories = top_categorias.sort_values(ascending=False)
```

```python
top_categorias
```

```python
# Top 3 productos por género en cada grupo etario

agrupar = df.groupby(['gender', 'product_category'])['total_amount'].sum()
top3 = agrupar.groupby(level=[0], group_keys=False).nlargest(3)
top3_df = top3.reset_index()
top_combinations = top3_df.sort_values(by='total_amount', ascending=False)
```

```python
top_combinations
```
