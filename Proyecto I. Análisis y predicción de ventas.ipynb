```python
import pandas as pd
import numpy as np
```

```python
path = "/content/drive/MyDrive/Data Science - Skillnet/Ejercicios Core/BBDD/retail_sales_dataset.csv"
df = pd.read_csv(path)
```

```python
#Para visualizar la data
df.info()
```

```python
#Calcula el total de ventas por categoría de producto.

df.groupby('Product Category')['Total Amount'].sum()
```

```python
#Calcula el promedio de ventas diarias por categoría de producto.

df.groupby(['Date', 'Product Category'])['Total Amount'].mean()
```

```python
#Identifica las categorías de productos con mayores y menores ventas.

df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
```

```python
#Filtra los datos para mostrar solo las ventas de una categoría de producto específica.

df[df['Product Category'] == 'Beauty']['Total Amount']
```

```python
#Realiza operaciones de suma, resta, multiplicación y división en los datos para obtener estadísticas adicionales.

df['Total Amount'].sum()

#Calculando descuento aplicando una resta
df1 = df['Price after Discount'] = df['Price per Unit'] - 10

 #Usando multiplicacion para calcular precio Total.
df2 = df['Calculated Total'] = df['Quantity'] * df['Price per Unit']

#Usando división para calcular Precio promedio por unidad
df3 = df['Unit Price per Quantity'] = df['Total Amount'] / df['Quantity']
```
